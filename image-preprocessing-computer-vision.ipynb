# Image preprocessing using computer vision

Thresholding is both simple and effective method for image segmentation. We can apply as below for preprocessing noisy images:

1. Converting gray scale
2. Binary Thresholding
3. Adaptive Mean Thresholding
4. Adaptive Gaussian Thresholding


```python
import cv2
import numpy as np
import glob

plates = glob.glob("plate/img_1007.jpg")
processed = glob.glob("plate/processed1.jpg")
resized = glob.glob("plate/resized1.jpg")



def adaptiveThreshold(plates):
    for i, plate in enumerate(plates):
        img = cv2.imread(plate)

        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        cv2.imshow('gray', gray)
        
        # global thresholding
        # First type of Simple Thresholding is Binary Thresholding
        # After thresholding the image with this type of operator, we will
        # have image with only two values, 0 and 255.

        ret, thresh = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY)
        
        #Threshold value is the mean of neighbourhood area
        
        threshMean = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 10)
        
        # Threshold value is the weighted sum of neighbourhood values where weights are a gaussian window
        
        threshGauss = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 51, 27)
      
        cv2.imwrite("plate/processed1.jpg".format(i), threshGauss)

      


def resize(processed):
    for i, image in enumerate(processed):
        image = cv2.imread(image)

        ratio = 200.0 / image.shape[1]
        dim = (200, int(image.shape[0] * ratio))

        resizedCubic = cv2.resize(image, dim, interpolation=cv2.INTER_CUBIC)

        cv2.imwrite("plate/resized1.jpg".format(i), resizedCubic)


adaptiveThreshold(plates)
resize(processed)

```
